TREE

```python
class TreeNode:
    def __init__(self,value,children=[]):
        self.value = value
        self.children = children
    def isLeaf(self):
        return len(self.children) == 0
```

```python
a=TreeNode(7)#
b=TreeNode(8)#
c=TreeNode(6)#
d=TreeNode(5)#
e=TreeNode(4)#
f=TreeNode("Hello")#
g=TreeNode([1,2,3])#
h=TreeNode(3.14)#
i=TreeNode(b.value)#
j=TreeNode(f.value)
```

```python
f.children=[a,g]
a.children=[e,h,i]
b.children=[c,d,TreeNode("Hi")]
g.children=[b,j]
```

```python
def NLR_TreeTraversal(currNode):
    print(currNode.value)
    for child in currNode.children:
        if child.isLeaf():
            continue
        NLR_TreeTraversal(child)
    return
```

```python
NLR_TreeTraversal(f)
```

```python
def LRN_TreeTraversal(currNode):
    if currNode.isLeaf():
        print(currNode.value)
        return
    for child in currNode.children:
        LRN_TreeTraversal(child)
    print(currNode.value)
```

```python
LRN_TreeTraversal(f)
```

BINARY TREE

```python
class BinaryTreeNode:
    def __init__(self, key, left=None, right=None):
        self.key = key
        self.left = left
        self.right= right
    def isLeaf(self):
        return self.left==None and self.right==None
```

```python

def height(node):
    if node==None:
        return 0
    print(node.key)
    return 1+max(height(node.left), height(node.right))
def level(node,currentLevel):
    if node==None:
        return
    if node.left==None and node.right==None:
        print(currentLevel)
    level(node.left,currentLevel+1)
    level(node.right,currentLevel+1)
    
```

```python
a=BinaryTreeNode(3)
a.left=BinaryTreeNode(9)
b=BinaryTreeNode(20)
a.right=b
b.left=BinaryTreeNode(15)
b.right=BinaryTreeNode(7)
level(a,1)
```

```python

class BST:
    def __init__(self,root=None):
        self.root = root
    def insert(self,key):
        if self.root==None:
            self.root = BinaryTreeNode(key)
            return
        helper_insert(key,self.root)
        
def helper_insert(key,currNode):
    if key>currNode.key:
        if(currNode.right!=None):
            helper_insert(key,currNode.right)
        else:
            currNode.right=BinaryTreeNode(key)
    else:
        if (currNode.left != None):
            helper_insert(key,currNode.left)
        else:
            currNode.left=BinaryTreeNode(key)
```
